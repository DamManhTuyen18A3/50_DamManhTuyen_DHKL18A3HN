```python
#Câu 1
class NhanVien:
    def __init__(self, ma_nv, luong_cb):
        self.ma_nv = ma_nv
        self.luong_cb = luong_cb

    def tinh_luong(self):
        return self.luong_cb
class NhanVienTangCa(NhanVien):
    def __init__(self, ma_nv, luong_cb,):
        super().__init__(ma_nv, luong_cb)
        self.ma_nv = ma_nv
        luong_cb = luong_cb
        
    def tinh_luong(self):
        return self.luong_cb * 1.2
class NhanVienChuyenChuyen(NhanVien):
    def __init__(self, ma_nv, luong_cb):
        super().__init__(ma_nv, luong_cb)
        self.ma_nv = ma_nv
        luong_cb = luong_cb
    def tinh_luong(self):
        return self.luong_cb + 500000
if __name__ == "__main__":
    nv1 = NhanVien("NV001", 10000000)
    nv2 = NhanVienTangCa("NV002", 10000000)
    nv3 = NhanVienChuyenChuyen("NV003", 10000000)

    print(f"Lương nhân viên {nv1.ma_nv}: {nv1.tinh_luong()} VND")
    print(f"Lương nhân viên {nv2.ma_nv}: {nv2.tinh_luong()} VND")
    print(f"Lương nhân viên {nv3.ma_nv}: {nv3.tinh_luong()} VND")
```

```python
import numpy as np

san_luong = [10, 20, np.nan, 40, 50, np.nan]

arr = np.array(san_luong, dtype=float)

nan = np.isnan(arr)
mean_nan = np.nanmean(arr)
arr[nan] = mean_nan

print(arr)
san_luong_lon = arr[arr >9]
gia_tri_nho_nhat = np.min(san_luong_lon)
print("Giá trị nhỏ nhất trong sản lượng lớn hơn 9 là:", gia_tri_nho_nhat)
```

```python
import pandas as pd
class NhanSu:
    def __init__(self, filepath):
        self.filepath = filepath
    def loadfile(self):
        load_csv = pd.read_csv(self.filepath)
        return load_csv
filepath = "nhan_su.csv"
nhansu = NhanSu(filepath)
df = nhansu.loadfile()
print(df)
```

```python
mask = df['Ma_NV'].str.contains('Ns', case=False, na=False)
mask_nam = df['Nam_Cong_Tac']>=2021
ket_qua = df[mask & mask_nam]
print(ket_qua)
```

```python
import numpy as np
arr = df['Thanh_Tich'].to_numpy()
mask_na = np.isnan(arr)
mean_nan = np.nanmean(arr)
arr[mask_na] = mean_nan 
df['Thanh_Tich'] = arr
print(df)
```

```python
tong_hop = df.groupby('Phong_Ban')['Thanh_Tich'].sum()
print(tong_hop)
```

```python
pivot = df.pivot_table(index='Phong_Ban'
                       , columns='Nam_Cong_Tac',
                         values='Thanh_Tich', 
                         aggfunc='mean',
                         fill_value=0)
print(pivot)
```

```python
import sqlite3
conn = sqlite3.connect('he_thong.db')
sql = pivot.to_sql('he_thong', 
             conn,
             if_exists='replace',
             index=False)
read_sql = pd.read_sql('SELECT * FROM he_thong', conn)
print(read_sql.head())
conn.close()
```
